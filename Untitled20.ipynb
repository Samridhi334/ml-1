### üî¢ **Basic Concepts**

1. **What is a parameter?**
   A parameter is a configuration variable internal to the model which is estimated from the data during training. E.g., weights in linear regression.

2. **What is correlation?**
   Correlation measures the linear relationship between two variables. Values range from -1 (perfect negative) to 1 (perfect positive).

3. **What does negative correlation mean?**
   It means as one variable increases, the other decreases. Example: More exercise, less weight.

4. **Define Machine Learning. What are the main components in Machine Learning?**
   ML is a method of data analysis that automates model building.
   Components:

   * Data
   * Model
   * Algorithm
   * Loss function
   * Optimizer
   * Evaluation metrics

5. **What are continuous and categorical variables?**

   * Continuous: Numeric values (e.g., height, weight)
   * Categorical: Distinct categories (e.g., gender, city)

6. **What is causation? Explain difference between correlation and causation with an example.**
   Causation implies one variable causes a change in another.
   Correlation: Ice cream sales ‚Üë and drowning ‚Üë (not causal).
   Causation: Smoking ‚Üí Lung cancer.

7. **How does loss value help in determining whether the model is good or not?**
   Lower loss indicates better model performance. It quantifies error between predicted and actual values.

8. **What is an Optimizer? What are different types of optimizers?**
   An optimizer adjusts model parameters to reduce loss.
   Types:

   * SGD: Stochastic Gradient Descent
   * Adam: Adaptive Moment Estimation
   * RMSprop, Adagrad, etc.

### üß† **ML Workflow & Preprocessing**

9. **How do we handle categorical variables in Machine Learning?**
   Techniques:

   * Label Encoding
   * One-Hot Encoding
   * Binary Encoding

10. **What do you mean by training and testing a dataset?**
    Training set is used to train the model, testing set is used to evaluate its performance.

11. **What is a Test set?**
    A separate portion of the data used to assess how well the model generalizes to unseen data.

12. **How do we split data for model fitting in Python?**
    Using `train_test_split()` from `sklearn.model_selection`.

13. **Why do we perform EDA before fitting a model?**
    To understand the data, check distributions, find missing values, outliers, and correlations.

14. **What is feature scaling? How does it help?**
    Feature scaling brings all features to the same scale. It improves convergence and model performance, especially for distance-based algorithms.

15. **How do we perform scaling in Python?**
    Using `StandardScaler`, `MinMaxScaler`, etc. from `sklearn.preprocessing`.

16. **Explain data encoding.**
    Encoding transforms categorical variables into numerical form so ML models can process them.
